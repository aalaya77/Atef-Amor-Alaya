```python
creer_slide_nav(
    section_id="methode_construction",
    section_num="01",
    section_titre="Introduction",
    sous_titre="M√©thode de Construction : L'It√©ration <br>  <br> (Iterated Functions System IFS)",
    contenu="""

 <p>Parmi les dm√©thodes de construction de fractales, la m√©thode appel√©e 
 <span style="color: green; font-weight: bold;"> "Iterated Functions System" (IFS)</span> :
Il s'agit de partir d'une fonction holomorphe $f$ sur le plan complexe, un polyn√¥me par exemple, 
puis de considerer <span style="color: #e74c3c; font-weight: bold;">la suite des iter√©es</span> d'un nombre complexe donn√© par composition successive de $f$ 
<span style="color: #e74c3c; font-weight: bold;">(orbite) </span><p>
    
    <p><span style="color: #e74c3c; font-weight: bold;"> On se limite au cas o√π la fonction $f$
    est le polyn√¥me du second degr√©:</span>
    <p><span style="font-size: 20px;">‚û•</span> $f_c(z) := z^2 + c,$ avec $c\in \mathbb{C}$ es
    t un param√®tre. </span><p>

    <p>Ceci revient √†  consid√©rer la suite de nombres complexes:<p>
       
    <div style="background-color: #f0f7ff; padding: 10px; border: 2px dashed #2e86c1; border-radius: 15px; 
    text-align: center; margin: 12px 0;">
        <span style="font-size: 20px; color: #2c3e50;">
            $$z_0\in \mathbb{C},\;\;z_{n+1} = z_n^2 + c, n\geq .0$$
        </span>
    </div>

<div style="background-color: #2c3e50; padding: 20px; border: 2px dashed #2e86c1; border-radius:
15px; text-align: center; margin: 12px 0;">
        <span style="font-size: 18px; color: #f0f7ff;">
            Soit $z_0$ un point de $\mathbb{C}$. On d√©finit la suite $(z_n)_{n\geq0}$ par :
$z_{n+1} = f_c(z_n) =f_c^{n+1}(z_0)$.
On appelle cette suite l‚Äôorbite de $z_0$ sous l‚Äôaction de $p_c$ et on la note $\mathcal{O}^+(z_0).$
        </span>
    </div>

    <p>Le processus suit trois √©tapes fondamentales :</p>
    
    <div style="display: flex; flex-direction: column; gap: 10px;">
        <div style="background-color: #fcfcfc; padding: 10px; border-left: 5px solid #2e86c1;">
            <b>1. Initialisation :</b> On choisit un point de d√©part $z_0$ dans le plan complexe.
        </div>
        
        <div style="background-color: #fcfcfc; padding: 10px; border-left: 5px solid #2e86c1;">
            <b>2. It√©ration :</b> On applique la formule de mani√®re r√©p√©t√©e 
            ($z_0$ $&#10142;$ $z_1$ $&#10142;$  $z_2$ $&#10142;$ $z_3$ $&#10142;$ $\dots$).
        </div>
        
        <div style="background-color: #fcfcfc; padding: 10px; border-left: 5px solid #2e86c1;">
            <b>3. Test de divergence :</b> On observe si la suite reste born√©e ou s'√©chappe vers 
            l'infini (vitesse d'√©chappement).
        </div>
    </div>

    Consequence sur la couleur qu'on affectera au point $z_0$:
    
    <div style="margin-top: 15px; font-style: italic; color: #7f8c8d; font-size: 14px;">
        ‚û• C'est la vitesse √† laquelle les points s'√©chappent qui d√©termine la couleur de chaque pixel.
    </div>
    """,
    prev_id="proprietes_fractales"
)
```

```python
# --- On d√©finit le contenu avec le code source format√© ---
contenu_impl = r"""
<p>L'impl√©mentation repose sur une approche <b>vectoris√©e</b> avec NumPy pour maximiser les performances :</p>

<div style="background-color: #2d2d2d; color: #f8f8f2; padding: 15px; border-radius: 10px; font-family: 'Courier New', Courier, monospace; font-size: 0.85em; overflow-x: auto; line-height: 1.4;">
<span style="color: #66d9ef;">def</span> <span style="color: #a6e22e;">mandelbrot</span>(h, w, max_iter=<span style="color: #ae81ff;">50</span>):<br>
&nbsp;&nbsp;&nbsp;&nbsp;y, x = np.ogrid[-<span style="color: #ae81ff;">1.4</span>:<span style="color: #ae81ff;">1.4</span>:h*<span style="color: #ae81ff;">1j</span>, -<span style="color: #ae81ff;">2</span>:<span style="color: #ae81ff;">0.8</span>:w*<span style="color: #ae81ff;">1j</span>]<br>
&nbsp;&nbsp;&nbsp;&nbsp;c = x + y*<span style="color: #ae81ff;">1j</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;z = c<br>
&nbsp;&nbsp;&nbsp;&nbsp;divertime = np.zeros(z.shape, dtype=int)<br><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #f92672;">for</span> i <span style="color: #f92672;">in</span> <span style="color: #66d9ef;">range</span>(max_iter):<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;z = z**<span style="color: #ae81ff;">2</span> + c<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;diverge = z*np.conj(z) > <span style="color: #ae81ff;">4</span><br>            
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;escaping_now = diverge & (divertime == <span style="color: #ae81ff;">0</span>)<br>  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;divertime[escaping_now] = i<br>           
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;z[diverge] = <span style="color: #ae81ff;">2</span> <span style="color: #75715e;"># √âvite l'overflow</span><br><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #f92672;">return</span> divertime
</div>

<div style="margin-top: 15px; display: flex; gap: 10px;">
    <div style="flex: 1; border-left: 4px solid #2e86c1; padding-left: 10px; font-size: 0.9em;">
        <b>Vectorisation :</b> Le calcul est effectu√© sur toute la matrice <code>c</code> simultan√©ment, ce qui est bien plus rapide qu'une boucle sur chaque pixel.
    </div>
</div>
"""

# --- 1. Cr√©ation de la slide avec le code dans le contenu ---
creer_slide_nav(
    section_id="code_mandelbrot",
    section_num="02",
    section_titre="Analyse & Conception",
    sous_titre="Impl√©mentation de la fonction Mandelbrot",
    contenu=contenu_impl,
    prev_id="algorithme_echappement2"
)

# --- 2. Ex√©cution silencieuse pour afficher le r√©sultat visuel en dessous ---
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(h, w, max_iter=50):
    y, x = np.ogrid[-1.4:1.4:h*1j, -2:0.8:w*1j]
    c = x + y*1j
    z = c
    divertime = np.zeros(z.shape, dtype=int)
    for i in range(max_iter):
        z = z**2 + c
        diverge = z*np.conj(z) > 4            
        escaping_now = diverge & (divertime == 0)  
        divertime[escaping_now] = i           
        z[diverge] = 2 
    return divertime

plt.figure(figsize=(10, 4))
result = mandelbrot(400, 700)
plt.imshow(result, extent=[-2, 0.8, -1.4, 1.4], cmap='magma')
plt.axis('off')
plt.title("Rendu g√©n√©r√© par le code ci-dessus", fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import io
import base64

# --- 1. G√âN√âRATION DU GRAPHIQUE DE LA CARDIO√èDE ---
def get_cardioid_plot():
    t = np.linspace(0, 2*np.pi, 200)
    # √âquation param√©trique de la fronti√®re de la cardio√Øde de Mandelbrot
    # c = (1/2)exp(it) - (1/4)exp(2it)
    c_re = 0.5 * np.cos(t) - 0.25 * np.cos(2*t)
    c_im = 0.5 * np.sin(t) - 0.25 * np.sin(2*t)
    
    plt.figure(figsize=(5, 5), facecolor='none')
    plt.plot(c_re, c_im, color='#e74c3c', linewidth=2, label="Fronti√®re $\rho=1$")
    plt.fill(c_re, c_im, color='#fdf2f2', alpha=0.5)
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.title("Lieu des points fixes attractifs", fontsize=10)
    plt.axis('equal')
    plt.grid(True, linestyle='--', alpha=0.3)
    
    buf = io.BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight')
    plt.close()
    return base64.b64encode(buf.getvalue()).decode('utf-8')

card_img = get_cardioid_plot()

# --- 2. AFFICHAGE DE LA SLIDE ---
creer_slide_nav(
    section_id="analyse_cardioide",
    section_num="04",
    section_titre="Analyse Dynamique Profonde",
    sous_titre="La Cardio√Øde : Condition de Stabilit√©",
    contenu=f"""
    <div style="display: flex; gap: 20px; align-items: start;">
        <div style="flex: 1.2;">
            <p>Le bulbe principal de Mandelbrot est une <b>cardio√Øde</b>. Elle repr√©sente les param√®tres $c$ pour lesquels $f_c(z)$ poss√®de un <b>point fixe attractif</b>.</p>
            
            <div style="background: #f8f9fa; padding: 15px; border-radius: 10px; border-left: 5px solid #34495e;">
                <b style="color: #2c3e50;">D√©monstration rapide :</b><br>
                1. Point fixe : $z^2 + c = z \implies z^2 - z + c = 0$<br>
                2. Stabilit√© : $|f'_c(z)| < 1 \implies |2z| < 1$<br>
                3. √Ä la limite ($|2z|=1$), on pose $z = \\frac{1}{2}e^{{i\\theta}}$
            </div>
            
            <p style="margin-top: 10px;">En injectant $z$ dans l'√©quation du point fixe, on obtient le lieu g√©om√©trique :</p>
            <p style="text-align: center; font-size: 16px; color: #e74c3c;">
                $c = \\frac{1}{2}e^{{i\\theta}} - \\frac{1}{4}e^{{2i\\theta}}$
            </p>
        </div>
        
        <div style="flex: 0.8; text-align: center;">
            <img src="data:image/png;base64,{card_img}" style="width: 100%; border-radius: 10px;">
            <p style="font-size: 11px; color: #7f8c8d; margin-top: 5px;">Fronti√®re th√©orique de la zone de convergence</p>
        </div>
    </div>
    """,
    prev_id="mecanisme_transformation"
)
```

```python
creer_slide_nav(
    section_id="demonstration_mathematique_complete",
    section_num="04",
    section_titre="Analyse Dynamique Profonde",
    sous_titre="Analyse de l'application $f(z) = z^2$",
    contenu="""
    <div style="display: flex; gap: 20px; align-items: stretch;">
        
        <div style="flex: 1; display: flex; flex-direction: column; gap: 15px;">
            <div style="background: #fdf2f2; padding: 15px; border-radius: 10px; border-left: 5px solid #e74c3c;">
                <b style="color: #e74c3c;">1. Approche Polaire (Points & Cercles)</b><br>
                Soit $z = r e^{i\\theta}$, alors $f(z) = r^2 e^{i(2\\theta)}$.
                <ul style="font-size: 13px; margin-top: 8px;">
                    <li><b>Module :</b> $|w| = r^2$. Un cercle de rayon $r$ devient un cercle de rayon $r^2$.</li>
                    <li><b>Argument :</b> $\\arg(w) = 2\\theta$. L'espace est "enroul√©" deux fois sur lui-m√™me.</li>
                </ul>
            </div>
            
            <div style="background: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #2e7d32; font-size: 13px;">
                <b style="color: #2e7d32;">Cons√©quence Dynamique :</b><br>
                Si $r < 1$, l'it√©ration converge vers $0$.<br>
                Si $r > 1$, l'it√©ration diverge vers $\\infty$.<br>
                La fronti√®re critique est le cercle unit√© ($r=1$).
            </div>
        </div>

        <div style="flex: 1; background: #f0f7ff; padding: 15px; border-radius: 10px; border-left: 5px solid #2e86c1;">
            <b style="color: #2e86c1;">2. Approche Cart√©sienne (Droites)</b><br>
            En posant $z = x + iy$, on a $u = x^2 - y^2$ et $v = 2xy$.
            
            <p style="font-size: 13px; margin-top: 10px;"><b>Pour une droite verticale ($x = cste$) :</b></p>
            <p style="font-size: 14px; text-align: center; background: white; padding: 5px; border-radius: 5px;">
                $u = x^2 - \\left(\\frac{v}{2x}\\right)^2$
            </p>
            <p style="font-size: 13px;">On obtient une <b>parabole horizontale</b> s'ouvrant vers la gauche.</p>
            
            <p style="font-size: 13px; margin-top: 10px;"><b>Pour une droite horizontale ($y = cste$) :</b></p>
            <p style="font-size: 13px;">Le raisonnement sym√©trique m√®ne √† une parabole s'ouvrant vers la droite.</p>
        </div>
    </div>

    <div style="margin-top: 20px; padding: 10px; background: #2c3e50; color: white; border-radius: 8px; text-align: center; font-size: 15px;">
        <b>Bilan :</b> Ces d√©formations continues expliquent l'aspect "courb√©" et "auto-similaire" des fractales de Julia.
    </div>
    """,
    prev_id="dynamique_plotly_z2_large"
)
```
